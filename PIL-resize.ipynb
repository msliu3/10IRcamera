```python
import numpy as np
from PIL import Image,ImageFilter

from skimage import morphology,data,color
import matplotlib.pyplot as plt
import cv2

image = Image.open("./original.jpg")
width = image.width
height = image.height
print(width,height)
# image = image.resize((1600,1200),Image.BILINEAR)
for i in range(10):
    image = image.resize((image.width+width,image.height+height),Image.BILINEAR)
    
image.show()



img = cv2.imread('./enlarge-original.jpg',0)
ret,thresh1=cv2.threshold(img,80,1,cv2.THRESH_BINARY)
plt.imshow(thresh1,'gray')  
plt.title("THRESH_BINARY")  
plt.xticks([]),plt.yticks([])
plt.show()  

# image=color.rgb2gray(thresh1)
# image=1-image #反相

skeleton =morphology.skeletonize(thresh1)#thresh1
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax1.imshow(img, cmap=plt.cm.gray)#img
ax1.axis('off')
ax1.set_title('original', fontsize=20)
ax2.imshow(skeleton, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('skeleton', fontsize=20)
fig.tight_layout()
plt.show()
```
